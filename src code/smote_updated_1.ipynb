```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.over_sampling import SVMSMOTE
from imblearn.combine import SMOTEENN
from imblearn.combine import SMOTETomek

file_loc = '/content/drive/MyDrive/Colab Notebooks/'
output_loc = '/content/drive/MyDrive/Colab Notebooks/'
fn = ['w2vec']
```

```python

sm = SMOTE(random_state=42)
cou=0


for i in range(0,1):
    fname = file_loc+fn[i]+'_train_labelled.csv'
    df=np.genfromtxt(fname,delimiter=',')
    datan=df[:,0:-1]
    out=df[:,-1]
    X_res, y_res = sm.fit_resample(datan,out)
    y=y_res.reshape(-1,1)
    d=np.concatenate((X_res,y),axis=1)
    fname= output_loc + 'smote_'+fn[i]+'.csv'
    np.savetxt(fname,d, delimiter=',', fmt='%f')
```
